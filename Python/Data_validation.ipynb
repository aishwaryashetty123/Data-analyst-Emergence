# **Check for Duplicates**

```python
# Count how many rows (excluding the first occurrence) are duplicates in events table
count_of_duplicates = df_event.duplicated(subset=['customer_id', 'event_type', 'event_date', 'source'])
print(count_of_duplicates.sum())
```

```python
# Remove duplicates keeping only the first occurance
df_event_no_duplicates = df_event.drop_duplicates(subset=['customer_id', 'event_type', 'event_date', 'source'], keep='first')
```

```python
# Reset index after removing duplicate rows
df_event_no_duplicates.reset_index(drop=True, inplace=True)
df_event_no_duplicates.info()
```

```python
# Check for duplicates in subscription table
count_of_duplicates2 = df_sub.duplicated(subset=['customer_id', 'start_date', 'end_date', 'monthly_price', 'status'])
print(count_of_duplicates2.sum())
```

```python
# Remove duplicate rows keeping only the first occurance
df_sub_no_duplicates = df_sub.drop_duplicates(subset=['customer_id', 'start_date', 'end_date', 'monthly_price', 'status'], keep='first')
```

```python
# Reset index after removing duplicate rows
df_sub_no_duplicates.reset_index(drop=True, inplace=True)
df_sub_no_duplicates.info()
```

# **Handle missing values in customer table**

```python
# Fill missing 'segment' data with 'Unknown'
df_cust['segment'].fillna('Unknown', inplace=True)
df_cust.info()
```

```python
# Drop rows where signup date is missing
df_cust.dropna(subset=['signup_date'], inplace=True)
df_cust.info()
```

```python
# Reset index
df_cust.reset_index(drop=True, inplace=True)
df_cust.info()
```

# **Check for inconsistencies in data**

```python
# Filter for cancelled status AND inconsistent times
inconsistent_cancelled = df_sub_no_duplicates[
    (df_sub_no_duplicates['status'] == 'canceled') &
    (df_sub_no_duplicates['end_date'] < df_sub_no_duplicates['start_date'])]
print(inconsistent_cancelled)
```
